```python
valid_temp_range = (np.nan, 50)  # Replace 50 with the upper limit of valid temperatures

# Find the indices of outlier values
outlier_indices = np.where(np.logical_or(a2 < valid_temp_range[0], a2 > valid_temp_range[1]))

# Print the indices of outlier values
if len(outlier_indices[0]) > 0:
    print("Indices of outlier values:")
    for idx in range(len(outlier_indices[0])):
        month, week, day, temp_type = outlier_indices[0][idx], outlier_indices[1][idx], outlier_indices[2][idx], outlier_indices[3][idx]
        print(f"Month {month + 1}, Week {week + 1}, Day {day + 1}, Temp Type {temp_type}: Value {a2[month, week, day, temp_type]}")
else:
    print("No outlier values found in the dataset.")
```

```python
valid_temp_range = (np.nan, 50)  # Replace 50 with the upper limit of valid temperatures


outlier_indices = np.where(np.logical_or(a2 < valid_temp_range[0], a2 > valid_temp_range[1]))


for idx in range(len(outlier_indices[0])):
    month, week, day, temp_type = outlier_indices[0][idx], outlier_indices[1][idx], outlier_indices[2][idx], outlier_indices[3][idx]
    

    neighboring_values = a2[month, week, day]
    neighboring_values = np.delete(neighboring_values, temp_type)
    
   
    a2[month, week, day, temp_type] = np.median(neighboring_values)


print("Modified Temperature Data:")
print(a2)

```
